### Read the original JSON from dataset repo and parse for evaluation

```python
import json

with open('ori_pqau.json') as file:
    data = json.load(file)

qa_dataset = list(data.values())
parsed_sample = []
subject = "diabetes"
for item in qa_dataset:
    if subject.casefold() in item["QUESTION"].casefold():
        parsed_data = { "question": item["QUESTION"], "answer": item["LONG_ANSWER"] }
        parsed_sample.append(parsed_data)
    if len(parsed_sample) >= 10:
        break

print(parsed_sample)
```

### Write parsed sample to file

```python
with open("evaluation.jsonl", "a") as f:
    for sample in parsed_sample:
        f.write(json.dumps(sample) + "\n")
```
